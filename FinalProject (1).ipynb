```python
#Using the mean ratio and the pooled standard deviation, I created a funnel plot in order to determine if there was a 
#publication bias with any of our sources.
```

```python
wb = xlrd.open_workbook('Project Data Collection.xlsx', on_demand = True)
ws = workbook.sheet_by_name('Comparisons')

x1 = [1.00,1.05,0.97,0.98,0.95,0.86,0.94,1.01]
y1 = [0.6083,0.6213,0.6687,0.4788,0.5056,0.6255,0.3250,0.4977]

x2 = [0.77,1.01,0.96,0.95,0.99,1.01]
y2 = [1.3173,0.5729,0.8250,0.5858,0.7217,0.7109]

x3 = [1.00,0.95]
y3 = [0.2232,0.7333]

x4 = [1.06]
y4 = [0.9181]

x5 = [1.03,0.96,0.93,1.03,0.94]
y5 = [6.1784,3.3782,6.1227,8.8152,1.9407]

x6 = [0.94,1.50,1.04,1.03,0.97,1.00,0.88,1.00,0.90,0.98,0.94,0.88,0.95,0.85,0.98,\
      0.93,1.04,0.99,0.95,0.87,0.99,0.96,0.90,0.93,0.94,0.87,1.02,1.04,0.93,0.92,\
      0.88,0.97,0.98,0.96,0.88,0.80,0.90,1.00,0.94,1.01,0.89,0.98]
y6 = [0.3526,0.2299,0.5086,0.2678,0.4405,0.3298,0.2436,0.4036,0.3188,0.4300,0.4357,\
      0.4355,0.4257,0.4426,0.3985,0.0594,0.7501,0.4351,0.2378,0.4500,0.3932,0.1988,\
      0.2207,0.3814,0.2882,0.2396,0.5274,0.4193,0.3726,0.3312,0.9228,0.4051,0.4031,\
      0.4125,0.2950,0.3080,0.3802,0.3333,0.3590,0.3677,0.2855,0.2481]


#Create the funnel tringle
plt.plot([0.96,0.96],[0.01,10], 'k')
plt.plot([0.6,0.96],[0.01,10], 'k')
plt.plot([0.96,1.32],[10,0.01], 'k')

#Plot the brain sections
plt.scatter(x1,y1, color="red")
plt.scatter(x2,y2, color="darkorange")
plt.scatter(x3,y3, color="gold")
plt.scatter(x4,y4, color="limegreen")
plt.scatter(x5,y5, color="dodgerblue")
plt.scatter(x6,y6, color="orchid")

#Clean up the plot
plt.ylim(0.01,10)
plt.xlabel('Mean Ratio')
plt.ylabel('Log of Pooled SD')
plt.title('Publication Bias')

plt.legend(['Average Mean','Funnel Boundary','Funnel Boundary','Caudate','Putamen','Globus Pallidus','Thalamus','Temporal','Hippocampus'], loc='upper right')
plt.yscale('log')

fig1 = plt.gcf()
plt.show()
plt.draw
fig1.savefig('publication_bias.png', dpi=100)
```
